```python
'''This Project is about analysing the Birth rate of Australia, USA and Russia for a different time periods. The dataset 
is sourced from mortality.org website. This dataset contains information of male and female birth population for each year.'''

'''This dataset is collected from mortality.org, where most of countries population data is maintained. I have taken dataset
on birth population of three countries Australia, Russia and USA and combined the files into a single dataset.'''
# Read Text Files with Pandas using read_csv()

# importing pandas
import pandas as pd

# importing matplotlib library to plot data
import matplotlib.pyplot as plt 

# read text file into pandas DataFrame
#sep=',' is used since it is a comma seperated file
df = pd.read_csv("AusUSARus.txt",sep=',')
#Print dataframe
print(df)

```

```python
#Information about dataframe
df.info()
```

```python
#Checking for null values in all columns
df.isnull().sum()
```

```python
#Overall picture of dataframe
df.describe()
```

```python
#Creating a separate dataframe for USA population
df_USA = df[df['PopName']=='USA']
#Selecting required columns from the dataframe
df_USA_Births=df_USA[['Births','Sex','Year']]
# display USA country DataFrame
print(df_USA_Births)
```

```python
#Creating a separate dataframe for RUSSIA population
df_RUS = df[df['PopName']=='RUS']
#Selecting required columns from the dataframe
df_RUS_Births=df_RUS[['Births','Sex','Year']]
# display RUSSIA Country DataFrame
print(df_RUS_Births)
#As we can see for each year there are two rows, one for male and another for female count. We are separating the Russian 
#dataframe into two separate dataframes, one dataframe male population and another for female population.
male_R=df_RUS_Births[df_RUS_Births['Sex']=='m']

female_R=df_RUS_Births[df_RUS_Births['Sex']=='f']
```

```python
#Creating a separate dataframe for Australia population
df_AUS = df[df['PopName']=='AUS']
#Selecting required columns from the dataframe
df_AUS_Births=df_AUS[['Births','Sex','Year']]
# display Australia DataFrame
print(df_AUS_Births)
#As we can see for each year there are two rows, one for male and another for female count. We are separating the Australian
#dataframe into two separate dataframes, one dataframe male population and another for female population.
male_A=df_AUS_Births[df_AUS_Births['Sex']=='m']
female_A=df_AUS_Births[df_AUS_Births['Sex']=='f']
```

```python
#AUSTRALIA PLOT
#Plotting male population of Australia
ax=male_A.plot(x='Year',y='Births',alpha=0.5,label='male')
#Plotting female population of Australia on the same plot
female_A.plot(ax=ax,x='Year',y='Births',alpha=0.5,label='female')
#Labelling x and y axes of plot
plt.ylabel('Births')
plt.xlabel('Year')
#Giving title to the plot
plt.title('Australia Birth Rate Between 1860-2020')
#Function for plotting linear scale 
ax.ticklabel_format(useOffset=False, style='plain')
#Starts Y axis with zero reference
plt.ylim(ymin=0)
#Displays the plot
plt.show()
#USA PLOT
#As we can see for each year there are two rows, one for male and another for female count. We are separating the USA
#dataframe into two separate dataframes, one dataframe male population and another for female population.
male_U=df_USA_Births[df_USA_Births['Sex']=='m']

female_U=df_USA_Births[df_USA_Births['Sex']=='f']
#Plotting male population of USA
ax=male_U.plot(x='Year',y='Births',alpha=0.5,label='male')
#Plotting female population of USA on the same plot
female_U.plot(ax=ax,x='Year',y='Births',alpha=0.5,label='female')
#Labelling x and y axes of plot
plt.ylabel('Births')
plt.xlabel('Year')
#Giving title to the plot

plt.title('USA Birth Rate Between 1933-2021 ')
#Function for plotting linear scale
ax.ticklabel_format(useOffset=False, style='plain')
#Starts Y axis with zero reference
plt.ylim(ymin=0)
#Displays the plot
plt.show()

#RUSSIA PLOT
#Plotting male population of RUSSIA
ax=male_R.plot(x='Year',y='Births',alpha=0.5,label='male')
#Plotting female population of RUSSIA on the same plot
female_R.plot(ax=ax,x='Year',y='Births',alpha=0.5,label='female')
#Labelling x and y axes of plot
plt.ylabel('Births')
plt.xlabel('Year')
#Giving title to the plot
plt.title('Russia Birth Rate Between 1959-2014')
#Function for plotting linear scale
ax.ticklabel_format(useOffset=False, style='plain')
#Starts Y axis with zero reference
plt.ylim(ymin=0)
#Displays the plot
plt.show()
```

From the above graphs
In Australia, 
1)The no.of females are gradually decreasing compared to males.
2)From 1940s, the population growth is rapid and gradually increasing.

In USA,
1)Gender gap almost maintained equal.
2)The growth of population is increased ever since 1950s.

Incase of Russia,
1)In the initial years, the gender gap is compared to 1980-90s.Then same kind of trend is followed in later years.
2)The population was decresing from 1960s and began to stabilize 1980s and again population growth in later years.

Overall, for all three countries males birth rate was always more than females.

```python
#Summing the male and female population of USA for each year 
total_population_US=df_USA_Births['Births'].groupby(df_USA_Births['Year'])
#Plotting the USA population on line plot
ax=total_population_US.sum().plot(x='Year',y='Births',alpha=0.9,label='United States of America')

#Summing the male and female population of RUSSIA for each year 
total_population_RUS=df_RUS_Births['Births'].groupby(df_RUS_Births['Year'])
#Plotting the USA population on line plot
ax=total_population_RUS.sum().plot(ax=ax,x='Year',y='Births',alpha=0.9,label='Russia')

#Summing the male and female population of AUSTRALIA for each year
total_population_AUS=df_AUS_Births['Births'].groupby(df_AUS_Births['Year'])
#Plotting the USA population on line plot
total_population_AUS.sum().plot(ax=ax,x='Year',y='Births',alpha=0.9,label='Australia')

#Labeling for x and y axes
plt.ylabel('Births')
plt.xlabel('Year')
#Giving title for the plot
plt.title('Comparison between Birth rates of Australia, United States and Russia')
#Starts Y axis with zero reference
ax.ticklabel_format(useOffset=False, style='plain')
#Activates legend
plt.legend()
#Starts Y reference from 0
plt.ylim(ymin=0)
#Displays the plot
plt.show()
```

Note: In the above graphs, the graph for USA and Australia starts from 1933 and 1959 respectively. It is because the datset does not contain the data before years.
